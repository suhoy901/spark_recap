## 2. Story Writing Schema

```
from pyspark.sql.functions import *
from pyspark.sql.types import *

Storywriting_1050_schema = StructType([
  StructField("sem_id", LongType(), True),
  StructField("top_cors_id", LongType(), True),
  StructField("com_id", LongType(), True),
  StructField("g_seq", LongType(), True),
  StructField("app_ver", StringType(), True),
  StructField("serial", StringType(), True),
  StructField("mac_address", StringType(), True),
  StructField("manufacturer", StringType(), True),
  StructField("model", StringType(), True),
  StructField("submit_seq", LongType(), True),
  StructField("score", StringType(), True),
  StructField("study_time", LongType(), True),
  StructField("word_count", LongType(), True),
  StructField("word_max_count", LongType(), True),
  StructField("word_recomm_count", LongType(), True),
  StructField("word_min_count", LongType(), True),
  StructField("title", StringType(), True),
  StructField("std_email_to", StringType(), True),
  StructField("stdContent", StringType(), True),
  StructField("ginger_revisionText", StringType(), True),
  StructField("ginger_revisionData", ArrayType(
    StructType([
      StructField("tcidx", LongType(), True),
      StructField("revisionReplaceWord", StringType(), True),
      StructField("gCheckAnswer", StringType(), True),
      StructField("gSelectType", StringType(), True),
      StructField("beforeWord", StringType(), True),
      StructField("afterWord", StringType(), True),
    ])
  ), True),
  StructField("act_data", ArrayType(
    StructType([
      StructField("act_code", StringType(), True),
      StructField("act_name", StringType(), True),
      StructField("act_order_seq", StringType(), True),
      StructField("page_data", ArrayType(
        StructType([
          StructField("page_order_seq", StringType(), True),
          StructField("page_name", StringType(), True),
          StructField("model_answer", StringType(), True),
          StructField("std_answer1", StringType(), True),
          StructField("std_answer2", StringType(), True),
          StructField("std_answer3", StringType(), True)
        ])
      ), True)
    ])
  ), True)
])
```

```
# .load() -> .json() 
Storywriting_DF = spark.read.schema(picturecuedSpeaking_1049_schema).json("./mnt/cdpocmount/april/study_result/544/791/93673/1050/1422842/1422842_20200408104004000.txt")

Storywriting_DF.printSchema()
```
